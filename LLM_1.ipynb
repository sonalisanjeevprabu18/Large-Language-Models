```python
import sentencepiece as spm
from transformers import BertTokenizer

text = "Tokenization is important"
print("Text:", text)
print("-" * 30)

# 1. Word Tokenization
word_tokens = text.split()
print("Word Tokens:", word_tokens)
print("-" * 30)

# 2. Character Tokenization
char_tokens = list(text.replace(" ", ""))
print("Character Tokens:", char_tokens)
print("-" * 30)

# 3. Subword Tokenization (SentencePiece)
with open("text.txt", "w") as f:
    f.write(text)

spm.SentencePieceTrainer.train(
    input="text.txt",
    model_prefix="spm",
    vocab_size=30,
    model_type="bpe"
)

sp = spm.SentencePieceProcessor()
sp.load("spm.model")

subword_tokens = sp.encode(text, out_type=str)
print("Subword Tokens:", subword_tokens)
print("-" * 30)

# 4. WordPiece Tokenization (BERT)
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
wordpiece_tokens = tokenizer.tokenize(text)
print("WordPiece Tokens:", wordpiece_tokens)
```
